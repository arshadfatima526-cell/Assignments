```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("FINAL_USO_cleaned.csv")

# --------- PLOTS ---------
plt.figure(figsize=(14, 6))

# 1. Histogram of USO Closing Prices
plt.subplot(2, 3, 1)
df['USO_Close'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of USO Closing Prices")
plt.xlabel("USO Close Price")
plt.ylabel("Frequency")
print("Insight 1: Most USO closing prices fall within a common range, with few extreme values.")

# 2. Scatter Plot: USO Close vs Volume
plt.subplot(2, 3, 2)
plt.scatter(df['USO_Close'], df['USO_Volume'], alpha=0.5, color='purple')
plt.title("USO Close vs Trading Volume")
plt.xlabel("USO Close Price")
plt.ylabel("USO Volume")
print("Insight 2: No strong correlation between USO closing price and trading volume, "
      "though some volume spikes occur at mid-level prices.")

# 3. Line Chart of USO Closing Price Over Time
plt.subplot(2, 3, 3)
plt.plot(df['Date'], df['USO_Close'], color='green')
plt.title("USO Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("USO Close Price")
plt.xticks(rotation=45)
print("Insight 3: USO prices fluctuate over time, showing periodic volatility typical of energy ETFs.")

# 4. Bar Chart of Average Monthly USO Close
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')
monthly_avg = df.groupby('Month')['USO_Close'].mean().head(12)  # first year only
plt.subplot(2, 3, 4)
monthly_avg.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Average Monthly USO Close (First Year)")
plt.xlabel("Month")
plt.ylabel("Avg USO Close Price")
print("Insight 4: Monthly averages reveal seasonal or market-driven fluctuations in USO prices.")

# 5. Scatter Plot: USO Close vs S&P Open
plt.subplot(2, 3, 5)
plt.scatter(df['USO_Close'], df['SP_open'], alpha=0.5, color='red')
plt.title("USO Close vs S&P Opening Price")
plt.xlabel("USO Close")
plt.ylabel("S&P Open")
print("Insight 5: USO closing price and S&P opening price show a weak positive relationship.")

plt.tight_layout()
plt.show()
```
