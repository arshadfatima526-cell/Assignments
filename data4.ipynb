```python
import pandas as pd

# Load dataset
df = pd.read_csv("FINAL_USO_cleaned.csv")

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Correlation with target variable (USO_Close)
target_corr = correlation_matrix["USO_Close"].sort_values(ascending=False)

# Display all correlations
print("Correlation with USO_Close:\n", target_corr)

# Top 3 most related features (excluding USO_Close itself)
top3_features = target_corr.drop("USO_Close").head(3)
print("\nTop 3 features most related to USO_Close:\n", top3_features)
```
